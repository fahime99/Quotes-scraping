### Quotes Scraping

```python
# import libraries

import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://bookroo.com/quotes/anne-of-green-gables'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
quotes = []

for quote in soup.find_all('div', class_='css-eez2cp'):
    text = quote.text.strip()
    quotes.append(text)

print(quotes)
```

```python
characters = []

for character in soup.find_all('a', class_='chakra-link css-1s1o1fc'):
    text = character.text.strip()
    characters.append(text)

print(characters)
```

```python
ID=[]
for div in soup.find('div',class_='chakra-stack css-v5f6d'):
    id=div.get('id')
    if id != None:
      ID.append(id)
print(ID)
```

```python
f=1

for i in ID:
  characters=set()
  DIV = soup.find('div', {'id': i})
  divs = DIV.find_all("div")
  for div in divs:
    quote = div.find('div', class_='css-eez2cp')
    
    if quote is not None:
      text = quote.text.strip()
      print( f , '-' , text)

    character = div.find_all('a' , class_='chakra-link css-1s1o1fc')

    for c in character:
        if c is not None:
          text1 = c.text.strip()
          characters.add(text1)
          
  print(characters)
  print()
  f=f+1
```

```python
quotes_by_character = {}
f=1
for i in ID:
  DIV = soup.find('div', {'id': i})
  divs = DIV.find_all("div")
  for div in divs:
    quote = div.find('div', class_='css-eez2cp')
    
    if quote is not None:
      text = quote.text.strip()
      
      character_links = div.find_all('a' , class_='chakra-link css-1s1o1fc')
    
      characters = [link.text.strip() for link in character_links]
      
      for character in characters:
        if character not in quotes_by_character:
          quotes_by_character[character] = []
        quotes_by_character[character].append(text)

character_name = input('enter character name =')
character_name = character_name.title()
if character_name in quotes_by_character:
  print(f"Quotes by {character_name}:")
  for quote in quotes_by_character[character_name]:
    print(f)
    print(quote)
    print()
    f=f+1
else:
  print(f"No quotes found for {character_name}.")
```
